```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import *
from math import *
from sympy import *
```

```python
#Objective Function
n = int(input('Enter the degree of polynomial: '))
coeffs = symbols(' '.join(['A' + str(i) for i in range(n+1)]))
x = symbols('x')
f = sympify(sum(coeffs[i]*x**i for i in range(n+1)))
display(f)
g = sympify(input('Enter the target function: '))
display(g)
obj_func = f-g
display(obj_func)
```

```python
#Initial guesses
k = []
n_iter = int(input('Enter the no. of iterations: '))
for i in range(n_iter):
    guess = i*pi/(n_iter-1)
    k.append(guess)
display(k)
```

```python
def h(k,obj_func,x):
    res = 0
    for var in k:
        res += obj_func.subs(x,var)**2
    return res/n_iter
#Compute the grad(h)
expr = h(k,obj_func,x)
display(expr)
gradient = [diff(expr,coeff) for coeff in coeffs]
print('Gradient: ')
for i,grad in enumerate(gradient):
    print(f'd(expr)/d(A{i}) = ')
    display(grad)
```

```python
#Initialize
alpha = float(input('Enter the learning rate: '))
coeff_values = {}
for coeff in coeffs:
    if coeff != pi:
        coeff_values[coeff] = float(input(f'Enter the value for coefficient {coeff}: '))
pi_value = pi.evalf()
coeff_values[pi] = pi_value
f0 = expr.subs(coeff_values)
display(f0)
```
